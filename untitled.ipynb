```python
import pandas as pd

# Load data
df = pd.read_csv("/portfolio_data.csv")

# Ensure 'Date' is in datetime format and set it as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)




```

```python
# Total missing values per column
missing_values = df.isnull().sum()
print("üîç Missing Values:\n", missing_values)
```

```python
# Count unique values per column
unique_values = df.nunique()
print("\nüî¢ Unique Values:\n", unique_values)
```

```python
# Summary statistics using describe()
summary_stats = df.describe()
print("üìà Summary Statistics:\n", summary_stats)
```

```python
# Transpose to make rows = assets
print(df.describe().T)
```

```python
plt.figure(figsize=(14, 6))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Stocks')
plt.show()
```

```python
daily_returns = df.pct_change().dropna()

plt.figure(figsize=(12, 6))
for column in daily_returns.columns:
    sns.kdeplot(daily_returns[column], label=column, fill=True)

plt.title('Distribution of Daily Returns')
plt.xlabel('Return')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette='Set3')
plt.title('Price Distribution by Stock')
plt.ylabel('Price (USD)')
plt.show()
```

```python
rolling_df = df.rolling(window=30).mean()

plt.figure(figsize=(14, 6))
for column in rolling_df.columns:
    plt.plot(rolling_df.index, rolling_df[column], label=column)

plt.title('30-Day Rolling Average of Stock Prices')
plt.xlabel('Date')
plt.ylabel('Rolling Average Price')
plt.legend()
plt.show()
```

```python
cumulative_returns = (1 + daily_returns).cumprod()

plt.figure(figsize=(14, 6))
for column in cumulative_returns.columns:
    plt.plot(cumulative_returns.index, cumulative_returns[column], label=column)

plt.title('Cumulative Returns of Each Stock')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
for col in daily_returns.columns:
    sns.histplot(daily_returns[col], kde=True, label=col, bins=50, element='step', stat='density')
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
cumulative_returns = (1 + daily_returns).cumprod()
cumulative_returns.plot.area(figsize=(14, 6), colormap='tab20c', alpha=0.8)
plt.title('Cumulative Returns Area Plot')
plt.ylabel('Cumulative Return')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```
